```python
import timeit

import matplotlib.pyplot as plt
import pandas as pd

iter_ls = [10, 50, 100, 500, 1000, 1500, 2000, 3000, 4000, 5000, 6000, 10000]



def time_complex(n, func):
    t = []
    for j in n:
        ls = [1] * j
        start = timeit.default_timer()
        x = func(ls)
        end = timeit.default_timer()
        t.append(end - start)
    return t


def time_visual(t, n, title):
    df = pd.DataFrame(list(zip(n, t)), columns=['len', 'time'])
    print(df)

    plt.plot(n, t, "--p")
    plt.title(title)
    plt.xlabel("len")
    plt.ylabel("time")
    plt.show()


def r(a):
    n = len(a)
    x = [None] * n
    for i in range(n):
        x[n - 1 - i] = a[i]
    return x


time_visual(time_complex(iter_ls, r), iter_ls, "main")


def d(a):
    n = len(a)
    for i in range(n // 2):
        a[i], a[n - 1 - i] = a[n - 1 - i], a[i]
    return a


time_visual(time_complex(iter_ls, d), iter_ls, "optimal")


# ----------------------------------------------------------------

def avg(s):
    n = len(s)
    a = [0] * n
    for j in range(n):
        total = 0
        for i in range(j + 1):
            total += s[i]
        a[j] = total / (j + 1)
    return a


time_visual(time_complex(iter_ls, avg), iter_ls, "main")


def avg2(s):
    n = len(s)
    a = [0] * n
    total = 0
    for j in range(n):
        total += s[j]
        a[j] = total / (j + 1)
    return a


time_visual(time_complex(iter_ls, avg2), iter_ls, "optimal")
```
